```python
import numpy as np
import pandas as pd
import os
import datetime as dt
```

```python
path_dir = r'C:\Users\WIN10\Documents\Hodu\2022_10_KTSDE\data\news'
file_list = os.listdir(path_dir)
```

```python
df_concat = pd.read_excel(path_dir+'/'+file_list[0])
df_concat = df_concat[['일자','제목']]

for file in file_list[1:]:
    df_new = pd.read_excel(path_dir+'/'+file)
    df_new = df_new[['일자','제목']]
    df_concat = pd.concat([df_concat, df_new], ignore_index=True)
```

```python
df_concat.info()
```

```python
df_concat['일자'] = pd.to_datetime(df_concat['일자'], format = '%Y%m%d')
df_concat
```

```python
df_concat = df_concat.sort_values(by=['일자'], axis = 0)
df_concat = df_concat.reset_index(drop=True)
```

```python
kospi_df = pd.read_excel(r'C:\Users\WIN10\Documents\Hodu\2022_10_KTSDE\data\kospi.xlsx')
```

```python
kospi_df['일자'] =pd.to_datetime(kospi_df['일자'])
```

```python
kospi_df['일자'] = kospi_df['일자'].dt.date
```

```python
kospi_df['일자'] = kospi_df.sort_values(by=['일자'], axis = 0)
kospi_df = kospi_df.reset_index(drop=True)
```

```python
kospi_df['일자'] =pd.to_datetime(kospi_df['일자'])
```

```python
kospi_df.info()
```

```python
df_merge = pd.merge(kospi_df, df_concat, how='left', left_on = '일자', right_on = '일자')
df_merge
```
